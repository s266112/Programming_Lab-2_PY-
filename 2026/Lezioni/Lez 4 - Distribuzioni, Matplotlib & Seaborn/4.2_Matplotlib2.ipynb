### Diagrammi a barre: `bar()`

````python
colori_occhi = np.array(['nero', 'blu', 'verde', 'marrone'])
freq_assoluta = np.array([1,4,5,10])
# Aggiorniamo i colori delle barre per corrispondere al colore degli occhi
colori_barre = ['black', 'blue', 'green', 'brown']

# Creazione del diagramma a barre con colori corrispondenti
plt.figure(figsize=(10, 6))
bars = plt.bar(colori_occhi, freq_assoluta, color=colori_barre)

# Aggiunta di una legenda per identificare i colori
plt.legend(bars, colori_occhi)

# Aggiunta del titolo e delle etichette agli assi
plt.title('Frequenza Assoluta del Colore degli Occhi')
plt.xlabel('Colore Occhi')
plt.ylabel('Frequenza Assoluta')

# Visualizzazione del diagramma
plt.show()

````

# Diagrammi circolari: `pie()`


* `pie()`: crea un grafico a torta.
* Parametri comuni:
    * `y` - Etichette o dati da visualizzare (colonna o lista).
    * `labels` - Etichette da mostrare per ogni fetta.
    * `colors` - Lista di colori da usare per le varie sezioni.
    * `autopct` - Mostra i valori numerici dentro ogni fetta (es. `'%1.1f%%'`).
    * `startangle` - Angolo iniziale del grafico (es. 90° per ruotarlo).
    * `explode` - Lista di valori per "staccare" le fette dal centro.
* Sintassi:  
  `plt.pie(df['data'], labels=df['category'], autopct='%1.1f%%', startangle=90, colors=['red', 'green', 'blue', 'yellow'], explode=(0.1, 0, 0, 0))`
* Il grafico mostra come le singole parti contribuiscono al totale.








````python
# Dati per il diagramma circolare degli occhi
labels = ['Nero', 'Azzurro', 'Verde', 'Marrone']
sizes = [1, 4, 5, 10]  # Frequenza assoluta per ogni colore di occhi
colors = ['black', 'blue', 'green', 'brown']

# Creazione del diagramma circolare
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Aggiunta del titolo
plt.title('Distribuzione dei Colori degli Occhi')

# Mostra il diagramma circolare
plt.show()
````

````python
# Dati per il diagramma circolare delle stelle
labels_stelle = ['1 Stella', '2 Stelle', '3 Stelle', '4 Stelle', '5 Stelle']
sizes_stelle = [1, 2, 4, 6, 7]  # Frequenza assoluta per ogni categoria di stelle
colors_stelle_gold = [  '#b5b17a','#b5a642','#d4af37', '#ffc700', 'gold']

# Creazione del diagramma circolare
plt.figure(figsize=(8, 8))
plt.pie(sizes_stelle, labels=labels_stelle, colors=colors_stelle_gold, autopct='%1.1f%%', startangle=140)

# Aggiunta del titolo
plt.title('Distribuzione delle Valutazioni a Stelle')

# Mostra il diagramma circolare
plt.show()
````

## Istogrammi

È un particolare diagramma a barre dove raggruppo un insieme di dati in un numero di rettangoli.

* `hist()`: crea un istogramma.
* Parametri comuni:
    * `x` - i dati da plottare
    * `bins` - numero di intervalli (o bordi degli intervalli)
    * `range` - intervallo minimo e massimo dei dati da includere
* Sintassi: `hist(x, bins=None, range=None)`

### Informazioni Generali

* Serve a capire la distribuzione di dati numerici continui mostrando la frequenza dei valori in intervalli specifici (bins).
* Può aiutare a rivelare pattern come asimmetrie, distribuzioni normali o bimodalità.
* L’asse x è diviso in intervalli (bin) di uguale dimensione, e l’altezza di ogni barra rappresenta la frequenza dei valori in quell’intervallo.

* Differenza con i grafici a barre:
    * Gli istogrammi mostrano **dati continui**, i grafici a barre **categorie discrete**.
    * L’asse x di un istogramma rappresenta gli intervalli di una variabile continua, mentre nei bar chart rappresenta le categorie.
    * Gli istogrammi sono usati per visualizzare **la distribuzione** di una variabile continua; i bar chart per **confrontare categorie**.


````python
dati = [
    1.55,
    1.653, 1.654, 1.655, 1.657,
    1.751, 1.752, 1.753, 1.754, 1.755, 1.756, 1.757,
    1.854, 1.855, 1.856, 1.857, 1.858, 1.859,
    1.951, 1.952
]

plt.figure(figsize=(8,5))
plt.hist(dati, bins=200, range=(1.50, 2.00), edgecolor='black')

plt.title('Distribuzione delle Altezze')
plt.xlabel('Altezza')
plt.ylabel('Frequenza')

plt.show()
````

````python
# genero k numeri casuali
k = 200
media = 0
varianza = 1
n = np.random.normal(media, varianza, k)

# bins sono il numero di rettangoli in cui dividere i dati, provate a variarlo
plt.hist(n, bins=8, linewidth=0.5, edgecolor="white")
plt.title("Histogram")
plt.show()
````

## Scatter plot (grafico a dispersione)
 Mostra una collezione di punti su un piano cartesiano. Ogni punto rappresenta una coppia di valori (x, y) e serve a visualizzare la relazione tra due variabili numeriche.



* `scatter()` - crea un diagramma a dispersione tra x e y.
* Parametri comuni:
    * `x`, `y` - Nomi delle colonne da usare per gli assi.
    * `s` - Dimensione di ogni punto (valore singolo o array).
    * `c` - Colore di ogni punto (singolo colore o array).
    * `marker` - Stile del marcatore (es. `o`, `^`, `*`, ecc.).
    * `alpha` - Trasparenza dei punti, da 0 (trasparente) a 1 (opaco).
* Sintassi:  
  `plt.scatter(df['x'], df['y'], s=df['size'], c=df['color'], alpha=0.5, marker='o')`
* I grafici a dispersione servono a confrontare categorie o gruppi di dati.

### Esempio

## Subplots

* `subplots()`: Crea più grafici all'interno di una singola figura.  
* Restituisce due oggetti:  
    1. `fig` - rappresenta l'intera finestra della figura che conterrà tutti i sottografi.  
    2. `ax` - può essere un singolo oggetto asse (per un solo grafico) o un array di assi (per più grafici).  
* Argomenti:
    * `nrows` e `ncols` determinano rispettivamente il numero di righe e colonne di sottografi.
    * `sharex` e `sharey` controllano se gli assi x o y devono essere condivisi tra i sottografi.
* Sintassi: `fig, ax = plt.subplots(nrows, ncols)`
* Esempio: `fig, ax = plt.subplots(1,2)` - Crea una figura con 1 riga e 2 colonne di grafici.


````python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crea 2 sottotrame (1 riga, 2 colonne)
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Primo subplot
ax[0].plot(x, y1)
ax[0].set_title("Seno")

# Secondo subplot
ax[1].plot(x, y2)
ax[1].set_title("Coseno")

plt.tight_layout()
plt.show()
````
