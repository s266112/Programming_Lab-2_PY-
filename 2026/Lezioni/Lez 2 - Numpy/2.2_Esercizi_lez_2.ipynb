### **3. Seleziona il numero più vicino a 0.5 in un array casuale**
Genera un array 10 x 3 con numeri casuali nell'intervallo [0,1].
Per ogni riga, seleziona il numero più vicino a 0.5. Per farlo usa il fancy indexing.

```python
import numpy as np

# 1. Genero array 10x3 con numeri casuali in [0,1)
array = np.random.rand(10, 3)

# 2. Calcolo la distanza assoluta da 0.5
distanza = np.abs(array - 0.5)

# 3. Trovo l'indice di colonna del minimo per ogni riga
indici_colonna = np.argmin(distanza, axis=1)

# 4. Seleziono i valori corrispondenti (fancy indexing)
valori_scelti = array[np.arange(10), indici_colonna]

# 5. Stampo il tutto
print("Array originale:\n", array, "\n")
print("Indici dei valori più vicini a 0.5:", indici_colonna, "\n")
print("Valori più vicini a 0.5 per ogni riga:\n", valori_scelti)
```

```python
import numpy as np

# 1. Creo il vettore stipendi e calcolo il costo totale del personale
stipendi = np.array ([50000, 105250, 55000, 89000])        # Vettore stipendi
costo_del_personale = np.sum(stipendi)                     # Costo del personale


# 2. Aumento del 15% per il CEO (stipendio 105250) con interi (per vedere l'errore)
stipendi_CEO_aumentato = stipendi.copy()                        # Copia per non modificare l'originale
stipendi_CEO_aumentato[1] = stipendi_CEO_aumentato[1] * 1.15    # 105250 * 1.15 = 121037.5 → viene troncato a 121037


# 3. Il 115% di 105250 è 121037.50? E' presente nell'array? Se no perchè?
valore_atteso = 121037.50
presente = valore_atteso in stipendi_CEO_aumentato

# 4. Ricreare l'array con dtype=float e ripetere l'aumento del CEO ottenendo esattamente 121037.50
stipendi_float = np.array([50000, 105250, 55000, 89000], dtype=float)
stipendi_float[1] = stipendi_float[1] * 1.15                    # Ora diventa 121037.5


# 5. Aumento del 20 % al dipendente con stipendio più basso (50000)
indice_min = np.argmin(stipendi_float)                          # Indice del valore minimo (0)
stipendi_float[indice_min] = stipendi_float[indice_min] * 1.20  # 50000 * 1.20 = 60000


# 6. Aumento del 10 % agli altri due dipendenti
for i in range(len(stipendi_float)):
    if i != 1 and i != indice_min:                              # i != 1 → non CEO, i != indice_min → non il più basso
        stipendi_float[i] = stipendi_float[i] * 1.10


# 7. Nuovo costo totale e differenza
totale_finale = np.sum(stipendi_float)
aumento_totale = totale_finale - costo_del_personale


# Print finali
print ("PUNTO 1:")
print("Vettore degli stipendi iniziali: ", stipendi)
print("Costo totale del personale: ", costo_del_personale, "\n")

print ("PUNTO 2:")
print("Stipendi dopo l'aumento del CEO (con interi):", stipendi_CEO_aumentato)
print("Il valore del CEO è diventato:", stipendi_CEO_aumentato[1], "\n")

print ("PUNTO 3:")
print(f"Valore atteso: {valore_atteso}")
print("È presente nell'array?", presente)
print("Motivo: L'array è di tipo intero, quindi la moltiplicazione per 1.15 ha prodotto 121037.5 ma è stato troncato a 121037. Il valore decimale non può essere rappresentato in un array intero." , "\n")

print("PUNTO 4")
print("Stipendi con float dopo aumento CEO:", stipendi_float)
print("Il valore del CEO è ora esattamente:", stipendi_float[1], "\n")

print("PUNTO 5")
print("Indice del dipendente con stipendio più basso:", indice_min)
print("Vettore degli stipendi dopo l'aumento (20 %) dell'impiegata:", stipendi_float, "\n")


print("PUNTO 6")
print("Vettore dopo l'aumento (10 %) degli altri dipendenti:", stipendi_float, "\n")

print("PUNTO 7")
print("Costo totale del personale dopo tutti gli aumenti:", totale_finale)
print("Aumento totale subito dall'azienda:", aumento_totale)
```

```python
import numpy as np

# 1. Generazione e salvataggio del file (se non esiste gia).
#    Simula la creazione del file "emissioni.txt" con 200 valori casuali tra 4 e 20
file_name = "emissioni.txt"
co2_emissioni = np.round(np.random.uniform(4, 20, 200), 2)

# Nome del file di output
output_file = "emissioni.txt"  # Salva il file nella cartella di lavoro corrente
np.savetxt(output_file, co2_emissioni, fmt="%.2f")


# 2. Carico il file
co2 = np.loadtxt('emissioni.txt')


# 3. Calcoli
soglia = 18                                 # Soglia di 18 tonnellate
co2_sopra_soglia = co2[co2 > soglia]        # Filtro i valori sopra la soglia
num_sopra = len(co2_sopra_soglia)           # Calcolo quanti valori sono sopra la soglia
media_sopra = np.mean(co2_sopra_soglia)     # Calcolo la media delle emissioni sopra la soglia


somma_sopra = np.sum(co2_sopra_soglia)      # Somma delle emissioni sopra la soglia
somma_totale = np.sum(co2)                  # Somma totale di tutte le emissioni
frazione = somma_sopra / somma_totale       # Calcolo la frazione

# Print finali
print(f"Numero di valori sopra {soglia} tonnellate: {num_sopra}")
print(f"Media di questi valori: {media_sopra:.2f}\n")
print(f"Somma sopra soglia: {somma_sopra:.2f}")
print(f"Somma totale: {somma_totale:.2f}")
print(f"Frazione di emissione totale sopra la soglia: {frazione:.4f}")
```
