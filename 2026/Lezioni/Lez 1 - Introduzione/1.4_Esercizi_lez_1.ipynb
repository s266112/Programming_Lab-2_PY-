### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

# ESEMPIO 1: CUBI DEI NUMERI DA 0 A 9
Voglio creare una lista che contenga i cubi dei numeri interi da 0 a 9 (inclusi).


```python
# Con Ciclo for
cubi = []
for i in range (10):
    cubi.append(i**3)
print(cubi)
```

```python
# Con list comprehesion
cubi =[i**3 for i in range(10)]
print(cubi)
```

### ***Cosa fa?:***:
- (i**3) eleva i alla terza potenza
- **range(10)** genera i numeri da 0 a 9.

# ESEMPIO 2: LETTERE MAIUSCOLE
Prendo una lista di parole, ad esempio ["ciao", "sono", "alessio"], e trasformo ogni parola nella sua versione **maiuscola**

```python
# Con Ciclo for
parole = ["ciao", "sono", "alessio"]
maiuscole = []
for parola in parole:
    maiuscole.append(parola.upper())
print (maiuscole)
```

```python
# Con list comprehesion
parole = ["ciao", "sono", "alessio"]
maiuscole = [parola.upper() for parola in parole]
print(maiuscole)
```

### ***Cosa fa?:***:
- ***.upper()*** è un metodo delle stringhe che restituisce la stringa in maiuscolo.
- La list comprehesion applica ***parola.upper()*** a ogni elemento della lista originale

# ESEMPIO 3: NUMERI DIVISIBILI PER 100 
Voglio creare una lista contenente tutti i numeri compresi tra 0 e 1000 (estremi inclusi) che sono **divisibili per 100**.

```python
# Con ciclo for
multipli_100 = []
for n in range(0, 1001):
    if n % 100 == 0:
        multipli_100.append(n)
print(multipli_100)
```

```python
# con list comprehesion
multipli_100 = [n for n in range(0, 1001) if n % 100 == 0]
print(multipli_100)
```

### ***Cosa fa?:***:
- **range(0, 1001)** genera i numeri da 0 a 1000.
- La condizione **if n % 100 == 0** seleziona solo i numeri il cui resto della divisione per 100 è zero, cioè i multipli di 100.
- La list comprehesion include la condizione dopo il **for**. 

### **2. Vettore di Numeri Primi**

1. **Crea un vettore** contenente tutti i numeri primi compresi tra **0 e 10**  
   _(Puoi scriverli direttamente nell'array)._

2. **Conta quanti numeri ci sono nel vettore** utilizzando la funzione `len()`.  **Ottieni lo stesso numero** accedendo all'attributo `.size` del vettore.

3. Quale pensi sia il tipo di dato (`dtype`) del vettore?  Prova a rispondere senza eseguire il codice e Verifica la tua risposta accedendo all'attributo `.dtype` del vettore

4. Scrivi l'array usando una list comprehension che controlla che i numeri siano primi. 

```python
# 1: Creo il vettore contenente i numeri primi da 0 a 10

import numpy as np

primi = np.array([2, 3, 5, 7])  # 0 e 1 non sono primi
print (primi)
```

```python
# 2: Conteggio degli elementi

# Con len()
print("Lunghezza con len():", len(primi)) # Output 4

# Con .size
print("Lunghezza con .size:", primi.size) # Output 4
```

```python
# 3: Tipo di dato (dtype)

print("Tipo di dato:", primi.dtype) # Output: int32 o int64
```

```python
# 4: Creazione dell'array con list comprehesion verificando se il numero è primo

def è_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

primi_lc = np.array([x for x in range(11) if è_primo(x)])
print(primi_lc)
```

### **3.  Operazioni su array**
Crea un array **a**, 1D (senza digitarlo esplicitamente), e genera due nuovi array **b**, contenente una sottostringa a piacere e **c** che è il reverse dell'array **a** (esempio [1,2,3] diventa [3,2,1]) .  Dividi l'array **a** per l'array **c**. Fai la stessa cosa per una lista.

# CON UN ARRAY NUMPY

```python
import numpy as np

# 1. Creo array a
a = np.arange(1, 6)   # [1, 2, 3, 4, 5]
print("a =", a)

# 2. Creo array b (slicing) – ad esempio i primi 3 elementi
b = a[:3]
print("b =", b)

# 3. Creo c come reverse di a
c = a[::-1]           # [5, 4, 3, 2, 1]
print("c =", c)

# 4. Divisione elemento per elemento
divisione = a / c
print("a / c =", divisione)
```

# CON UNA LISTA PYTHON

```python
# 1. CreO lista a
lista_a = list(range(1, 6))   # [1, 2, 3, 4, 5]
print("lista_a =", lista_a)

# 2. Slicing per b
lista_b = lista_a[:3]
print("lista_b =", lista_b)

# 3. Reverse
lista_c = lista_a[::-1]
print("lista_c =", lista_c)

# 4. Divisione elemento per elemento (con list comprehension)
divisione_lista = [lista_a[i] / lista_c[i] for i in range(len(lista_a))]
print("lista_a / lista_c =", divisione_lista)
```
