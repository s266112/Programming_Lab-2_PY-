# Concetti base di statistica

La **Statistica** è il metodo di studio dei fenomeni collettivi, i.e. di una **popolazione** (di persone ma anche di oggetti) e della loro intrinseca variabilità.

La statistica è la scienza che studia modelli, metodi e algoritmi per la raccolta, l'esplorazione, l'integrazione, l'analisi, la predizione e l'interpretazione di dati.

La statistica si pone lo scopo di stimare i parametri che descrivono la dinamica aleatoria a partire da dati, ossia da osservazioni ripetute del sistema reale, che il modello stocastico si propone di approssimare.

I dati vengono raccolti sull'intera popolazione  (rilevazione esaustiva) o su un **campione** (rilevazione parziale) è un sottoinsieme della popolazione.

Uno studente di un liceo scientifico può far parte di diverse popolazioni, secondo il contesto in cui è osservato:
1.  è un elemento della popolazione degli alunni frequentanti la sua stessa scuola,
2.  fa parte della popolazione dei cittadini residenti a una certa data in un dato comune,
3.  appartiene alla popolazione dei giovani della sua stessa fascia di età.

L'insieme formato da soli 5 studenti per ogni classe è un esempio di campione della popolazione di una scuola.


Un dato (statistico) è il risultato della misurazione/osservazione di una qualche caratteristica sugli elementi di una popolazione. La caratteristica si chiama **variabile**.

Possiamo raccogliere i dati in questo laboratiorio delle altezze e dei colori degli occhi dei vostri occhi. Altezza e colore degli occhi sono le caratteristiche (variabili) che voglio studiare.

Le caratteristiche possono essere qualitative (il colore degli occhi), quantitative discrete (età), quantitative continue (l'altezza).

Di che tipo sono i dati che raccoglierete?

Con il termine **modalità** si indicano i modi in cui una certa variabile si può manifestare in un collettivo, ad esempio "Blu, Marrone" sono modalità della variabile colore degli occhi,  i valori dell’età sono modalità della variabile Età.


La **frequenza (assoluta)** è il numero di volte in cui è stata osservata una modalità.

La **frequenza relativa** è il rapporto tra la frequenza assoluta e numero totale dei dati.

La **frequenza percentuale** è la frequenza relativa moltiplicata per 100.

# Distribuzioni

I dati raccolti possono essere classificati attraverso distribuzioni.
La distribuzione è una rappresentazione del modo in cui le diverse modalità/valori di un carattere si distribuiscono.

Le distribuzioni possono essere discrete o continue.

## Distribuzioni (di probabilità) per valori discreti (finiti)

Definite su un insieme a valori discreti.

Lo spazio degli stati S, è composto da un numero discreto di modalità/valori (e.g. i diversi colori degli occhi).

Ad ogni elemento è associata una probabilità di essere pescato, $p(x), x \in S$. Queste probabilità devono sommare a $1$.

Esempio di una sacca con 20 palline di colori diversi (1 nera, 4 blu, 5 verdi, 10 marroni).

$p(\text{nera}) = \frac{1}{20}, \quad p(\text{blu}) = \frac{4}{20}, \quad p(\text{verde}) = \frac{5}{20}, \quad p(\text{marrone}) = \frac{10}{20}$

$p(\text{nera}) = 0.05, \quad p(\text{blu}) = 0.2, \quad p(\text{verde}) = 0.25, \quad p(\text{marrone}) = 0.5$


1/20+4/20+5/20+10/20=1


````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import pandas as pd
import seaborn as sns
````

### Moda, media, mediana

*   La **Moda** è definita come il valore che ha la frequenza più alta

La media e la mediana si possono calcolare solo su  distribuzioni di un carattere quantitativo oppure qualitativo ordinabile.

*   La **Media (aritmetica)** è quel valore che corrisponde alla somma di tutti i valori diviso il numero dei valori stessi.
*   La **Mediana** è quel valore/modalità assunto dalle unità statistiche che si trovano nel mezzo della distribuzione

Come si calcola la mediana:

*   Si ordinano gli n dati in modo crescente
*   se il numero di dati è dispari la mediana corrisponde al valore centrale, ovvero al valore che occupa la posizione (n+1)/2
*   se il numero n di dati è pari, la mediana è stimata utilizzando i due valori che occupano le posizioni n/2 e n/2 +1 (generalmente si sceglie la loro media aritmetica se il carattere è quantitativo).



````python
colori = np.array(['18', '19', '20', '21', '22'])
vettore_valori=np.array([18,19,19,19,19,19,19,19,19,19,20,20,20,20,20,20,21,21,21,22])
n = len(vettore_valori)
media = np.sum(vettore_valori)/n
if n%2==0:  # se n è pari
  mediana = (vettore_valori[n//2-1] + vettore_valori[n//2])/2.  # considerando l'indice che parte da zero quindi devo "slittare" di uno indietro
else:
  mediana = vettore_valori[(n-1)//2] # (n+1)/2 -1 , considerando l'indice che parte da zero quindi devo "slittare di uno indietro"
print('media:', media)
print('mediana:', mediana)

#esistono i metodi in numpy
print('media:', np.mean(vettore_valori))
print('mediana:', np.median(vettore_valori))
````

## Distribuzioni (di probabilità) per valori continui

Definite su un insieme continuo. Ad esempio un intervallo dei numeri reali.

### Distribuzione Uniforme
La probabilità ha un valore costante in un dato intervallo [a,b] e zero fuori dall'intervallo. Si indica di solito con U(a,b) dove a e b sono gli estremi dell'intervallo. Il valore costante è pari a 1/(b-a).

````python
X = st.uniform(0,1)
X2 = st.uniform(2,2)
fig, ax = plt.subplots()
xs = np.linspace(-0.1, 4.1, 200)
ax.plot(xs, X.pdf(xs),label='U[0,1]')
ax.plot(xs, X2.pdf(xs),label='U[2,4]')
ax.set_xlabel('$x$')
ax.set_ylabel('$p(x)$');
plt.title('distribuzione uniforme continua')
ax.legend()
````



### Distribuzione Normale (o di Gauss)
Serve per descrivere variabili casuali a valori reali che tendono a concentrarsi attorno a un singolo valor medio (teorema del limite centrale).

Ha una forma come di una campana rovesciata dove la punta della campana corrisponde proprio al valore medio.

La curva è simmetrica rispetto al valore medio

Moltissime modalità attorno a voi hanno una distribuzione normale. Ad esempio l'altezza ed il peso delle persone, od i voti scolastici.

La distribuzione normale dipende da due parametri, la media $\mu$ e la varianza $\sigma^2$, ed è indicata tradizionalmente con: $N(\mu, \sigma^2)$,

La varianza è una misura della dispersione dei valori di una variabile statistica rispetto alla sua media, cioè è la misura di quanto i valori si discostano (quadraticamente) dalla media aritmetica.

La varianza di un insieme di dati $(x_1, x_2, \dots, x_n)$ è definita come:
$
\text{Var}(X) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2
$

dove $n$ è il numero totale di elementi, $x_i$ è il valore dell'i-esima osservazione, $\mu$ è la media dei valori, cioè $\mu = \frac{1}{n} \sum_{i=1}^{n} x_i $.

La sua radice quadrata, $\sigma$, si chiama deviazione standard.

````python
mu = 175.0
sigma = 10.0
X = st.norm(loc=mu, scale=sigma)

fig, ax = plt.subplots()
xs = np.linspace(mu - 6.0 * sigma, mu + 6.0 * sigma, 100)
ax.plot(xs, X.pdf(xs))
ax.axvline(np.mean(xs), color='r', linestyle='--', label="Mean")
ax.axvline(np.median(xs), color='g', linestyle='-', label="Median")
ax.set_xlabel('$x$')
ax.set_ylabel('$p(x)$')
plt.xlim([100, 250])
ax.legend()
````

````python
mu = 175.0
sigma = 20.0
X = st.norm(loc=mu, scale=sigma)

fig, ax = plt.subplots()
xs = np.linspace(mu - 6.0 * sigma, mu + 6.0 * sigma, 100)
ax.plot(xs, X.pdf(xs))
ax.axvline(np.mean(xs), color='r', linestyle='--', label="Mean")
ax.axvline(np.median(xs), color='g', linestyle='-', label="Median")
ax.set_xlabel('$x$')
ax.set_ylabel('$p(x)$')
plt.xlim([100, 250])
ax.legend()
````

Distribuzione normale standardizzata: $N(0, 1)$ significa che la media è zero e la varianza 1.

````python
# Estrarre un singolo numero casuale da una distribuzione normale con media 0 e deviazione standard 1
np.random.normal(loc=0.0, scale=1.0, size=None)
````

````python
vettore_probabilita = [0.05,0.1,0.2,0.3,0.35]

campione_discreto_10 = np.random.choice(a=[1,2,3,4,5], size=10, replace=True, p=vettore_probabilita)
plt.hist(campione_discreto_10, [1, 2, 3, 4, 5, 6], linewidth=0.5, edgecolor="white")
plt.xticks(np.arange(1, 6))
plt.title('DISCRETA: 10 campioni')
plt.show()

campione_discreto_100 = np.random.choice(a=[1,2,3,4,5], size=100, replace=True, p=vettore_probabilita)
plt.hist(campione_discreto_100, [1, 2, 3, 4, 5, 6], linewidth=0.5, edgecolor="white")
plt.xticks(np.arange(1, 6))
plt.title('DISCRETA: 100 campioni')
plt.show()

campione_discreto_10K = np.random.choice(a=[1,2,3,4,5], size=10000, replace=True, p=vettore_probabilita)
plt.hist(campione_discreto_10K, [1, 2, 3, 4, 5, 6], linewidth=0.5, edgecolor="white")
plt.xticks(np.arange(1, 6))
plt.title('DISCRETA: 10000 campioni')
plt.show()
````

````python
def plot_metrics(campione, titolo, nret=10):
  df = pd.DataFrame(campione, columns = ['samples'])

  f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)})
  mean=df['samples'].mean()
  median=df['samples'].median()


  plt.title(titolo)
  sns.boxplot(data=df, x="samples", ax=ax_box)
  ax_box.axvline(mean, color='r', linestyle='--')
  ax_box.axvline(median, color='g', linestyle='-')

  sns.histplot(data=df, x="samples", ax=ax_hist, bins=nret)
  ax_hist.axvline(mean, color='r', linestyle='--', label="Mean")
  ax_hist.axvline(median, color='g', linestyle='-', label="Median")

  ax_hist.legend()

  ax_box.set(xlabel='')

  plt.show()
  plt.close()
````

**Distribuzione Uniforme**

````python
# 10 campioni
print("Media: " + str(np.mean(campione_uniforme_10)))
print("Mediana: " + str(np.median(campione_uniforme_10)))

plot_metrics(campione_uniforme_10,'UNIFORME: 10 campioni')

# 100 campioni
print("\nMedia: " + str(np.mean(campione_uniforme_100)))
print("Mediana: " + str(np.median(campione_uniforme_100)))
plot_metrics(campione_uniforme_100,'UNIFORME: 100 campioni')

# 10k campioni
print("\nMedia: " + str(np.mean(campione_uniforme_10K)))
print("Mediana: " + str(np.median(campione_uniforme_10K)))
plot_metrics(campione_uniforme_10K, 'UNIFORME: 10K campioni')
````

**Distribuzione Gaussiana**

````python
# 10 campioni
print("Media: " + str(np.mean(campione_normale_10)))
print("Mediana: " + str(np.median(campione_normale_10)))
plot_metrics(campione_normale_10,'NORMALE: 10 campioni')

# 100 campioni
print("Media: " + str(np.mean(campione_normale_100)))
print("Mediana: " + str(np.median(campione_normale_100)))
plot_metrics(campione_normale_100,'NORMALE: 100 campioni')

# 10k campioni
print("Media: " + str(np.mean(campione_normale_10K)))
print("Mediana: " + str(np.median(campione_normale_10K)))
plot_metrics(campione_normale_10K,'NORMALE: 10K campioni')
````

**Osservazione 1**: Media e Mediana tendono ad assumere lo stesso valore all'aumentare del numero di dati generati presi in considerazione.

**Osservazione 2**: Media e Mediana tendono ad avvicinarsi al valore della media della distribuzione all'aumentare del numero di dati generati presi in considerazione.

````python
campione_gaussiana_10K_std1 = np.random.normal(loc=2, scale=1, size=10000)
print("Varianza: " + str(np.var(campione_gaussiana_10K_std1)))
print("Massimo valore generato :" + str(max(campione_gaussiana_10K_std1)))
print("Minimo valore generato :" + str(min(campione_gaussiana_10K_std1)))
plot_metrics(campione_gaussiana_10K_std1,'NORMALE: std 1')

campione_gaussiana_10K_std10 = np.random.normal(loc=2, scale=10, size=10000)
print("\nVarianza: " + str(np.var(campione_gaussiana_10K_std10)))
print("Massimo valore generato :" + str(max(campione_gaussiana_10K_std10)))
print("Minimo valore generato :" + str(min(campione_gaussiana_10K_std10)))
plot_metrics(campione_gaussiana_10K_std10,'NORMALE: std 10')
````

### Distribuzione asimmetrica

Distribuzione asimmetrica con coda di probabilità molto ampia. Qui media, mediana e moda differiscono di molto.

Un esempio è la distribuzione esponenziale, che descrive un processo di decadimento in cui la probabilità che un evento accada entro un intervallo di tempo è proporzionale al tempo stesso.

Esempi: durata di vita di un componente elettronico, tempo di attesa in una coda, gli intervalli di tempo tra due riparazioni di un macchinario

````python
campione_esponenziale = np.random.exponential(scale=1.,size=10000)
# il parametro scale è l'inverso del tasso di decadimento
campione_esponenziale_round = np.array([round(el, 1) for el in campione_esponenziale])

plot_metrics(campione_esponenziale,'ESPONENZIALE')
````

In una distribuzione esponenziale, la media e la mediana hanno valori diversi; la mediana (valore centrale è inferiore alla media).
